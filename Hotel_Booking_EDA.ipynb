```python
plt.figure(figsize=(8,4))
ax1=sns.countplot(x="hotel", hue="is_canceled", data=df )
# legend_labels,_ = ax1.get_legend_handles_labels()
# ax1.legend(bbox_to_anchor=(11,22))
plt.title("Resort and City hotel Reservation", size =20)
ax1.set_ylim(0,70000)
plt.xlabel("Hotels")
plt.ylabel("Reservations")
plt.legend(['Not canceled', 'Canceled'])
plt.show()
```

```python
plt.figure(figsize=(8,4))
ax2=sns.countplot(x='customer_type', hue='is_canceled',data=df)
plt.title("Reservations on the basis of the costumer type")
plt.xlabel("Customer Type")
sns.color_palette('rocket_r',as_cmap=True)
plt.ylabel("Reservations")
plt.legend(["Not canceled","canceled"])
ax2.set_ylim(0,60000)
plt.show()
```

```python
df['month'] =df['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
ax2=sns.countplot(x='month', hue='is_canceled',data=df , palette ="dark")
plt.title("Reservations on the basis of arrival months in a year")
plt.xlabel("Arrival Month")
plt.ylabel("Reservations")

plt.xticks(rotation=43)
plt.legend(["Not canceled","canceled"])
# ax2.set_ylim(0,10000)
plt.show()
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)*100
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)*100
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(12,5))
plt.title("Average daily rate in city and resort hotels ")

plt.plot(city_hotel.index, city_hotel['adr'], label="city hotel", color="y" )

plt.plot(resort_hotel.index, resort_hotel['adr'] , label="resort hotel", color ='c')

plt.xticks(rotation=40)
plt.legend()
plt.show()

```

```python
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date',inplace=True)

not_cancelled_data = df[df['is_canceled']==0]
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date',inplace=True)

plt.figure(figsize=(20,6))
plt.title('Average Daily Rate',size=20)
plt.ylabel("Reservations",size=20)
plt.xlabel("Dates",size=20)
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'], label='Not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'],label="Cancelled")
plt.legend(title="Reservations")
plt.show()
```

```python
cancelled_df_adr = cancelled_df_adr[(cancelled_df_adr['reservation_status_date']>'2016') & (cancelled_df_adr['reservation_status_date']<'2017-05')]
not_cancelled_df_adr = not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date']>'2016') & (not_cancelled_df_adr['reservation_status_date']<'2017-05')]

plt.figure(figsize=(20,6))
plt.title('Average Daily Rate',size=25)

plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'], label='Not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'],label="Cancelled")
plt.legend(title="Reservations")
plt.ylabel("Reservations",size=20)
plt.xlabel("Dates",size=20)
# plt.fill_between(not_cancelled_df_adr,cancelled_df_adr,alpha=0.25)
plt.show()
```
